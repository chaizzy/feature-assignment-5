```python
df5= pd.DataFrame({
    'Age':[25,42,37,19,31,28] ,
    'Soft drink Preference' :['Coke', 'Pepsi','Mountain dew','Coke','Pepsi','Coke']
})
```

```python
df5
```

```python
Average_age = df5.groupby('Soft drink Preference')['Age'].mean().to_dict()
```

```python
Average_age
```

```python
df5['Average_age']=df5['Soft drink Preference'].map(Average_age)
```

```python
df5
```

```python
# so we can see that 
# for average age of 24 preference is coke
# for average age of 37 preference is mountain dew
# for average age of 36 preference is pepsi
```
